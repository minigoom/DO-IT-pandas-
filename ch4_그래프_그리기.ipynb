# 4.2 matplotlib  라이브러리 자유자재로 사용하기

## 기초 그래프 그리기

```python
# 1. tips 데이터 집합을 불러와 변수 tips 에 저장합니다
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

```python
print(tips)
```

```python
# 2. 데이터프레임의 열 데이터 분포와 빈도를 살펴보는 용도의 그래프이다
# 변수를 한개만 사용하면 일변량그래프라고 부른다
```

```python
# 3. 기본틀 fig  를 마련하고 그래프 격자(axes1)을 구성해보자
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
```

```python
# 4. hist 매서드에 total_bill 열을 전달하면 히스토그램이 만들어진다
# x 축의 간격을 bins 로 정할 수 있다.
axes1.hist(tips['total_bill'], bins=10)
axes1.set_title('Histogram of Total bill')
axes1.set_xlabel("Frequency")
axes1.set_ylabel("Total Bill")

fig
```

```python
# 5. 산점도 그래프를 그려보자
# 변수가 2개를 사용하고 '이변량그래프'이다.
```

```python
# 6. 기본 틀과 그래프 격자를 만들고 scatter 매서드에 total_bill, tips 열을 전달하면서 산점도 그래프가 나타난ㄷ
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs TIps')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel("Tip")
```

```python
# 7. 박스 그래프를 그려보자
# - 이산형 변수와 연속형 변수를 함께 사용하는 그래프이다
```

```python
# 8. boxplot 매서드를 사용하면 박스 그래프를 그릴 수 있습니다.
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex']=='Female']['tip'],
              tips[tips['sex']=='Male']['tip']],
              labels=['Female','Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by sex')
```

## 다변량 그래프 그리기

```python
# 3개 이상의 변수를 사용한 그래프는 '다변량 그래프' 라고 부른다
# 다변량 그래프가 필요한 이유는?
# - 지불금액과 팁만을 사용하여 산점도 그래프를 그렸습니다. 만약 여기에서 성별을 추가하여 산점도 그래프를 표현하려면 어떻게 해야할까요?
```

```python

# 1. 앞에서 실습한 산점도 그래프에 성별을 새 변수로 추가해 보겠습니다
# 여성인 경우 0, 남성인 경우 1 로 반환한다
def record_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
# 2. record sex 매서드가 반환한 값(0,1 ) 을 데이터프레임에 추가하면 됩니다
tips['sex_color'] = tips['sex'].apply(record_sex)
```

```python
print(tips)
# 마지막 항에 추가된것을 확인 할 수 있다.
```

```python
# 3. 그래프를 그리는 김에 테이블당 인원수도 산점도 그래프에 추가해보자
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size']*10,
    c = tips['sex_color'],
    alpha = 0.5
)

axes1.set_title('total bill vs tip colored by sex and sized by size')
axes1.set_xlabel('total_bill')
axes1.set_ylabel('tip')
```

# 4.3 seaborn 라이브러리 자유자재로 사용하기

```python
# 1. 단변량 그래프 그리기 - 히스토그램
import seaborn as sns
 
tips = sns.load_dataset("tips")
```

```python
print(tips)
```

```python
# 2. seaborn 라이브러리로 히스토그램을 그리면 subplot, distplot 매서드를 사용하면 된다
# subplot 매서드로 기본틀을 만들고 distplot 매서드에 total_bill 열 데이터를 전달하면 히스토그램을 그릴 수 있다.

ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('total bill histogram with density plot')
```

```python
# 3. displot 매서드를 사용하면 히스토그램 밀집도를 그래프와 같이 그린다
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde = False)
ax.set_title('total bill histogram with density plot')
```

```python
# 4. 밀집도 그래프만 나타낼려면 hist 인지를 False 로 하면 됩니다
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist = False)
ax.set_title('total bill histogram with density plot')
```

```python
# 5. 실무환경에서는 데이터 시각화를 더 효율적으로 하기 위해서 여러 그래프를 한번데 출력하는 경우가 많다
# 밀집도 distplot 매서드에 rug 인자를 추가하여 그려보자

ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True)
ax.set_title('total bill histogram wit density and rug plot')
ax.set_xlabel('total bill')
```

```python
# 6. count 그래플 ㄹ 그려보자
ax = plt.subplots()
ax = sns.countplot('day',data=tips)
ax.set_title('count of days')
ax.set_xlabel('day of the week')
ax.set_ylabel('frequency')
```

## 다양한 종류의 이변량 그래프 그리기

```python
# 1. seaborn 라이브러리로 산점도 그래프 그리기
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('scatterplot of total bill and tip')
ax.set_xlabel('total bill')
ax.set_ylabel('tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y ='tip', data = tips, fit_reg = False)
ax.set_title('scatterplot of total bill and tip')
ax.set_xlabel('total bill')
ax.set_ylabel('tip')
```

```python
# 2. 산점도와 히스토그램을 한 번에 그려주는 jointplot 매서드를 보겠습니다
joint = sns.jointplot(x='total_bill', y = 'tip', data= tips)
joint.set_axis_labels(xlabel='Total bill', ylabel='Tip')
joint.fig.suptitle('Joint plot of Total bill and tips, fontsize =10, y = 1.0')
```

```python
# 3.육각형 산점도 그래프
hexbin = sns.jointplot(x='total_bill', y = 'tip', data= tips, kind='hex')
hexbin.set_axis_labels(xlabel='Total bill', ylabel='Tip')
hexbin.fig.suptitle('Joint plot of Total bill and tips, fontsize =10, y = 1.0')
```

```python
# 4. 이차원 밀집도 그리기
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'],
                 data2 = tips['tip'],
                 shade=True)
ax.set_title('kernel Density plot of Total bill and tip')
ax.set_xlabel('total bill')
ax.set_ylabel('tip')
```

```python
# 5. 바 그래프 그리기
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('bar plot of average total bill for time of day')
ax.set_ylabel('average total bill')
```

```python
# 6. 박스 그래프 그리기
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill',data=tips)
ax.set_title('boxplot of total bill by time of day')
ax.set_xlabel('time of day')
ax.set_ylabel('total bill')
```
